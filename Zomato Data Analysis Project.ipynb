# Zomato Data Analysis Project

```python
# import libraries.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
# Read Dataset

df = pd.read_csv('Zomato data .csv')
print(df)
```

```python
df
```

# Convert the Data type of column - rate.

```python
def handleRate(value) :
    value = str(value).split('/')
    value = value[0]
    return float(value)

df['rate'] = df['rate'].apply(handleRate)
print(df.head())
```

```python
df
```

```python
# Check missing value in dataset.

df.info()
```

# Q1. Type of returant.

```python
df.head()
```

```python
sns.countplot(x=df['listed_in(type)'],palette='Set2')
plt.xlabel("Type of retsturant")
```

# Conclusion - Majority of the resturant falls in dinning category.

```python
df.head()
```

```python
grouped_data = df.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes': grouped_data})
plt.plot(result, c ="green", marker="o")
plt.xlabel("Type of restaurant", c="red", size=15)
plt.ylabel("votes", c="red", size=15)
```

# Conclusion - Dinning Resturants has Recieved maximum votes.

```python
df.head()
```

```python
plt.hist(df['rate'],bins =5)
plt.title("ratings distribution")
plt.show()
```

# Conclusion - The majority resturants received  ratings from 3.5 to 4

## Average Order Spending by couples.

```python
df.head()
```

```python
couple_data = df['approx_cost(for two people)']
sns.countplot(x=couple_data, palette='Set1')
```

# Conclusion - The majority fo couples preferr resturents with an approximate cost of 300 rupees

## Which mode receives maximum rating.

```python
df.head()
```

```python
plt.figure(figsize = (6,6))
sns.boxplot(x= 'online_order', y = 'rate', data = df, palette='Set2')
```

# Conclusion - offline order recevied lower rating in comparison to online order.

```python
pivot_table = df.pivot_table(index='listed_in(type)', columns='online_order', aggfunc ='size', fill_value=0)
sns.heatmap(pivot_table, annot=True, cmap='YlGnBu',fmt='d')
plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("Listed In (Type)")
plt.show()
```

# Conclusion - Dinnig resturants primarily accept offline orders, whereas cafes primarily receive online orders. This suggests that clients preferred orders in person at restaurants, but prefer online ordering at cafes.
